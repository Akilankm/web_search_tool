# 00 — Notebook Gateway

This is the recommended starting point for the Product Evidence Harness. The notebooks are the user-facing gateway into the system.

Canonical documentation now lives in one place only: `docs/README.md`.


## Business purpose

The harness turns product web search into verified, auditable, product-coding-ready evidence. It does not blindly trust search rank and it does not treat every discovered URL as production-ready.

```text
Search -> evidence -> identity -> rendered-page relevance -> champion confirmation -> production handoff
```


## Choose your notebook

| Use case | Notebook | Output |
|---|---|---|
| Understand the repo | `00_notebook_gateway.ipynb` | Guided map |
| Prove one product end-to-end | `01_single_product_harness.ipynb` | Champion URL decision + concise review packet |
| Run many products | `02_batch_product_harness.ipynb` | Final CSV + review queue + row packets |
| Freeze a confirmed champion page offline | `03_offline_product_artifact.ipynb` | Local HTML + assets + validation |
| Review one row without opening many files | `04_review_artifact_reader.ipynb` | Notebook-rendered what/why/how decision view |


## Production handoff rule

Only hand off a URL automatically when all of these are true:

```text
product_url is not blank
production_url_ready = true
production_url_status = PRODUCTION_READY_EXACT_SCRAPABLE_BROWSER_URL
browser_openable = true
rendered_page_check_passed = true
highly_scrapable = true
exact_product_url_match = true
champion_confirmation.passed = true
champion_confirmation.success_count = champion_confirmation.required_successes
needs_review = false
```

Rows outside this filter are review-only. Hard-rejected candidates remain in `candidate_decisions.csv`; they are not selected evidence.


## Default row packet

```text
output/<row_id>/
├── final_row.csv
├── review_summary.md
├── review_decision.json
├── candidate_decisions.csv
└── product_coding_input.json
```

Open `review_summary.md` first. Use `candidate_decisions.csv` only to inspect selected and rejected candidates.


## Canonical docs

There is intentionally only one detailed docs file now:

```text
docs/README.md
```

Use it for the full operating contract, fields, statuses, and validation commands.


## Recommended route

```text
1. Read docs/README.md.
2. Open notebooks/01_single_product_harness.ipynb.
3. Run one product and inspect output/<row_id>/review_summary.md.
4. Open notebooks/04_review_artifact_reader.ipynb to render the row decision cleanly.
5. Open notebooks/02_batch_product_harness.ipynb for scale.
6. Use notebooks/03_offline_product_artifact.ipynb only if offline evidence is required.
```
