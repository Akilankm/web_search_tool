# 00 — Notebook Gateway

This is the recommended starting point for the Product Evidence Harness. The notebooks are the user-facing gateway into the system. Start here, choose the workflow, then run the matching notebook.

```text
Notebook first -> concise review artifacts -> decision contracts -> code internals only when needed
```


## Business purpose

The harness turns product web search into verified, auditable, product-coding-ready evidence. It does not blindly trust search rank and it does not treat every discovered URL as production-ready.

```mermaid
flowchart LR
    A[Product identity] --> B[Candidate discovery]
    B --> C[Tournament evaluation]
    C --> D[Evidence extraction]
    D --> E[Identity and scrapability checks]
    E --> F[Champion confirmation]
    F --> G[Production URL gate]
    G --> H[Concise review packet + coding evidence]
```


## Choose your notebook

| Use case | Notebook | Output |
|---|---|---|
| Understand the repo | `00_notebook_gateway.ipynb` | Guided map |
| Prove one product end-to-end | `01_single_product_harness.ipynb` | Concise review packet |
| Run many products | `02_batch_product_harness.ipynb` | Final CSV + review queue + row packets |
| Review one row without opening many files | `04_review_artifact_reader.ipynb` | Notebook-rendered what/why/how decision view |
| Freeze a confirmed champion page offline | `03_offline_product_artifact.ipynb` | Local HTML + assets + validation |

```mermaid
flowchart TD
    A[What do you need?] --> B{Workflow}
    B -->|One product demo| C[01 Single Product]
    B -->|Batch operation| D[02 Batch Harness]
    B -->|Review output row| E[04 Review Artifact Reader]
    B -->|Offline evidence after champion| F[03 Offline Artifact]
    C --> G[Open review_summary.md]
    D --> H[Inspect final CSV and row packets]
    E --> I[Show selected/rejected candidates]
    F --> J[Open offline/offline_page.html]
```


## Concise row packet

The default output is intentionally small and reviewer-friendly. Open `review_summary.md` first.

```text
output/<row_id>/
├── final_row.csv
├── review_summary.md
├── review_decision.json
├── candidate_decisions.csv
└── product_coding_input.json
```

`review_summary.md` answers: what was selected, why it was selected, how the decision was made, what the model/detectors contributed, what was rejected, and what a reviewer should do next.


## Production handoff rule

Only hand off a URL automatically when all of these are true:

```text
production_url_ready = true
production_url_status = PRODUCTION_READY_EXACT_SCRAPABLE_BROWSER_URL
champion_confirmation.passed = true
champion_confirmation.success_count = champion_confirmation.required_successes
needs_review = false
```

```mermaid
flowchart TD
    A[Selected URL] --> B{Production URL ready?}
    B -->|No| R[Review-only]
    B -->|Yes| C{Champion confirmation passed?}
    C -->|No| R
    C -->|Yes| D{Needs review?}
    D -->|Yes| R
    D -->|No| E[Automated handoff]
```
